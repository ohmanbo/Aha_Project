```python
import supervision as sv

dataset = sv.DetectionDataset.from_coco(
    images_directory_path="./datasets/images/",
    annotations_path="./datasets/COCO_motorcycle (pixel).json"
)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# 카테고리 목록 출력
print("클래스 목록:", dataset.classes)

# 전체 이미지 수
print("총 이미지 수:", len(dataset))

```

```python
def visualize_image_with_bboxes(start_index=0, num_images=5):
    for index in range(start_index, min(start_index + num_images, len(dataset))):
        image_path, image, annotations = dataset[index]
        print("이미지 경로:", image_path)

        # 마스크 시각화
        mask_annotator = sv.MaskAnnotator()
        box_annotator = sv.BoxAnnotator()

        annotated = mask_annotator.annotate(image.copy(), annotations)
        annotated = box_annotator.annotate(annotated, annotations)

        plt.figure(figsize=(10, 6))
        plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title(f"이미지: {image_path}")
        plt.show()
```

```python
visualize_image_with_bboxes(0)
```
